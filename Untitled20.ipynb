```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Define Parameters 

nx=50              #number of spatial points
nt=300             #number of time steps 
L=1.0              #length of the rode 
alpha=0.01         #thermal diffusivity 

dx=L/(nx-1)
dt=0.0005

#Stability Condition 
lambda_ = alpha * dt/dx**2

if lambda_ > 0.5:
    print("WARNING: Solution may be unstable! Reduce dt or increase dx.")
```

```python
#Initialize Temperature

u=np.zeros(nx)

#Initial condition: heat spike in the middle 
u[nx//2] = 100

#Boundary conditions (ends fixed at 0)
u[0]=0
u[-1]=0

#Store results for Visualization 
u_all = [u.copy()]
```

```python
#Time-Stepping Loop

for n in range(nt):
    u_new=u.copy()
    for i in range(1,nx-1):
        u_new[i]=u[i]+lambda_ * (u[i+1] - 2*u[i] + u[i-1])
    u=u_new.copy()

#Save at intervals for plotting
    if n % 50 == 0:
        u_all.append(u.copy())
```

```python
#Plot results 

x=np.linspace(0,L,nx)

plt.figure()
for i,temp in enumerate(u_all):
    plt.plot(x,temp,label=f"Step {i*50}")

plt.xlabel("Position along Rod")
plt.ylabel("Temperature")
plt.title("1D Heat Equation (Finite Difference Method)")
plt.legend()
plt.show()
```
